##### 3. Las columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates

##### 4. Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas


```python
# Paso 1. Creo una lista con las columnas que quiero modificar para luego hacer un "for" para cambiar los datos de esas columnas solamente
porcent = ["unemploymentrate", "totaltaxrate", "taxrevenue", "populationlaborforceparticipation", "outofpockethealthexpenditure", "grosstertiaryeducationenrollment", "grossprimaryeducationenrollment", "forestedarea", "cpichange", "agriculturalland"]

# Paso 2. Hago el for loop
for col in porcent:
    df_merge[col] = df_merge[col].str.replace("%", "")

# Paso 3. Verifico que los % han desaparecido
df_merge.head(2).T
```

##### 5. Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$"

```python
# Paso 1. Creo una lista con las columnas que quiero modificar para luego hacer un "for" para cambiar los datos de esas columnas solamente
dolar = ["gasolineprice", "gdp", "minimumwage"]

# Paso 2. Hago el for loop
for col in dolar:
    df_merge[col] = df_merge[col].str.replace("$", "")

# Paso 3. Verifico que los $ han desaparecido
df_merge.head(2).T
```
