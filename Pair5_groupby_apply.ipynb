Apply: Las columnas land_area, armed_forces_size, co2-emissions, gdp, population, urban_population, latitude, longitude, density, cpi son de tipo object pero en realidad deberían ser numéricas. Crea una función que nos permita cambiar el tipo de dato de estas columnas al que le corresponde

```python
#Estas columnas contienen valores en miles, hay que quitarles las comas y cambiarlos a interger
columnas = ["landarea","armedforcessize", "co2-emissions", "gdp", "population", "urban_population", "density"]

#Creamos la función para convertir los valores de las columnas que están todos en miles separados por comas. Para convertirlos a int tenemos que quitar la coma y convertirlos a interger
def cambiar_numerica(cadena): #Cadena se refiere al string de los valores de la columna
    try: 
        return int(cadena.replace(",", "")) 

    except:
        return np.nan

#itermos por cada una de las columnas y hacemos la modificaión definida en la función anterior "cambiar_numerica"
for columna in columnas:
    df[columna] = df[columna].apply(cambiar_numerica)
```

##### 5. ¿Cuál es la tasa de impuestos total máxima por densidad de población por continente? 
- Primero, necesitas calcular la tasa de impuestos total para cada continente. La tasa de impuestos total se encuentra en la columna "total_tax_rate"
- Luego, calcula la densidad de población para cada continente. La densidad de población se encuentra en la columna "density"
-  Divide la tasa de impuestos total entre la densidad de población para obtener la tasa de impuestos total por persona por kilómetro cuadrado
- Encuentra el país que tiene el valor más alto en esta nueva columna calculada

```python
df.info()
```

```python
#cambiamos totaltaxrate a float ya que es un objeto y no nos deja realizar la media

def cambiar_numerica4(cadena): 
    try:
        cadena = cadena.replace("%","") #primero le quitamos el porcentage y luego lo convertimos a float pq no lo podemos hacer todo de una vez
        return float(cadena) 
    except:
        return np.nan #añadimos esto pq en la columna hay varios nulos
    
#modificamos la columna de totaltaxtare
df["totaltaxrate"] = df["totaltaxrate"].apply(cambiar_numerica4)
```
